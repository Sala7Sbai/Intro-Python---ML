### Q1 : Sin function

````python
import matplotlib.pyplot as plt
import math
import numpy as np
x = np.linspace(0,15,200)
T= np.sin(x)  
plt.figure()
plt.plot(x,T,'r')
plt.xlabel('x')
plt.ylabel('y = sin(x)')
plt.title('La fonction sinus')
plt.show()
````

````python
import seaborn as sb

def plotiKolchi(df, col, plt_style = 'seaborn-ticks', color_palette = "bright"):

    dataGroupe=pd.DataFrame()
    dataGroupe['{} No'.format(col)] = df.groupby(col).size()
    dataGroupe['{} Proportion'.format(col)] = np.round(dataGroupe['{} No'.format(col)].divide(dataGroupe['{} No'.format(col)].sum())*100,0)
    
    print ('Nombre total :  {}:{}'.format(col,dataGroupe['{} No'.format(col)].sum()))
    
    plt.style.use(plt_style)
    sb.set_palette(sb.color_palette(color_palette))
    
    fig=plt.figure()
    
    aaa=dataGroupe['{} No'.format(col)].plot(kind='bar', title='{} Column'.format(col), figsize=(16,8), color=sb.color_palette())
    
    for p in aaa.patches: 
        aaa.annotate(np.round(p.get_height(),decimals=2),
                    (p.get_x()+p.get_width()/2., p.get_height()),
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points')

        
    
````

````python

plotiKolchi(dfc, 'sex')
plotiKolchi(dfc, 'survived')
plotiKolchi(dfc, 'age')
````
